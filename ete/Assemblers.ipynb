```python
with open('assembler.csv', 'w') as fw:
    fw.write("taxid,assembler"+"\n")
with open('assemblers-tax.csv', 'r') as f, open('assembler.csv', 'a') as fw:
    for line in f:
        start=line.find('[')+1
        end=line.find(']')
        taxid=line[start:end]
        row=line[start:]
        row=row[row.find('[')+1:]
        end=row.find(']')
        assembler=row[:end]
        fw.write(taxid+","+assembler+"\n")
        print ( taxid, assembler )
```
